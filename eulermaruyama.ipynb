```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eulermaruyama(X0,T,n,mu,sigma,RWcount):
    plt.figure(figsize=(20,10))
    for rw in range(RWcount):
        dt=T/n
        X=X0
        points=[]
        for item in range(n):
            X=X+mu*X*dt+sigma*X*np.sqrt(dt)*np.random.randn()
            points.append(X)
        plt.plot(np.linspace(0,T,n),points)
```

```python
eulermaruyama(0.1,1,365,0.,1,100)
```

```python
eulermaruyama(1,1,365,0,5,100)
```

```python
eulermaruyama(1,1,365,0,0,100)
```

```python
eulermaruyama(1,1,365,5,0,100)
```
